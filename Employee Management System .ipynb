```python
class Employee:
    def __init__(self, name,salary):
        self.name=name
        self.salary=salary
    def display(self):
        print("Name: ",self.name)
        print("Salary: ",self.salary)
class Manager(Employee):
    def __init__(self,name,salary,dept):
        super().__init__(name,salary)
        self.dept=dept
    def display(self):
        print("Name: ",self.name)
        print("Salary: ", self.salary)
        print("Department: ",self.dept)
m=Manager("Aftab",200000,"IT")
m.display()
```

**modifying Employees**

```python
employees=[
    Manager("Aftab",200000,"IT"),
    Manager("Ahmed",100000,"HR")
    
]

for emp in employees:
    emp.display()
    print()
```
